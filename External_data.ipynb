# Distance

```python
import imp
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
problem = imp.load_source('', 'problem.py')
X_df, y_array = problem.get_train_data()
```

```python
X_df['log_PAX'] = y_array
cols = X_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
X_df = X_df[cols]
X_df.head()
```

```python
mean_encode_dep = X_df.groupby('Departure')['log_PAX'].mean()
mean_encode_ar = X_df.groupby('Arrival')['log_PAX'].mean()
X_df.loc[:, 'dep_encod'] = X_df['Departure'].map(mean_encode_dep)
X_df.loc[:, 'ar_encod'] = X_df['Arrival'].map(mean_encode_ar)
X_df = X_df.drop(['log_PAX', 'WeeksToDeparture', 'std_wtd'], axis=1)
X_df.head()
```

```python
X_encoded = X_df
X_encoded.head()
```

```python
pd.unique(X_encoded.loc[:, 'Departure']).tolist()
```

```python
pd.unique(X_encoded.loc[:, 'Arrival']).tolist()
```

```python
col = ['Departure', 'Arrival', 'Distance']
dist = pd.read_csv('submissions/external_data/Dist.csv', sep=';', names = col)
dist.head()
```

```python
external_data = pd.read_csv('submissions/external_data/external_data.csv')
external_data = external_data.rename(
            columns={'Date': 'DateOfDeparture', 'AirPort': 'Arrival'})
external_data.head()
```

```python
X_ext = pd.merge(dist, external_data, how='left',
            left_on=['Arrival'],
            right_on=['Arrival'],
            sort=False, copy=False)
X_ext.head()
```

```python
#X_ext.to_csv('external_data_4.csv')
```

```python
X_encoded['dep_encod'] = np.round(X_encoded['dep_encod'], 2)
X_encoded['ar_encod'] = np.round(X_encoded['ar_encod'], 2)
X_encoded = X_encoded.drop(['DateOfDeparture'], axis=1)
X_encoded = X_encoded.drop_duplicates()
X_encoded.iloc[:,2].count()
```

```python
X_ext_2 = pd.merge( X_ext, X_encoded, how='left',
            left_on=['Departure', 'Arrival'],
            right_on=['Departure', 'Arrival'], sort=False)
X_ext_2.head()
```

```python
print(X_ext.iloc[:,4].count(), X_encoded.iloc[:,2].count(), X_ext_2.iloc[:,4].count())
```
